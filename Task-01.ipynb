```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
df = pd.read_csv("C:/Users/hp/Downloads/High_Value_Dataset__January_2024_20240718 - Copy.csv")

# Assuming the CSV has 'age' and 'gender' columns
# Bar chart for categorical variable (gender)
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(8, 5))
gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# Histogram for continuous variable (age)
plt.figure(figsize=(8, 5))
plt.hist(df['Age'], bins=10, color='green', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
